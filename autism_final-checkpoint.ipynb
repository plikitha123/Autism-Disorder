```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x=y)
plt.title('Class Distribution Before SMOTE')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
models = {
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss')
}
params = {
    'Random Forest': {'n_estimators': [100, 200], 'max_depth': [10, 20, None]},
    'Gradient Boosting': {'n_estimators': [100, 200], 'learning_rate': [0.05, 0.1], 'max_depth': [3, 5]},
    'Decision Tree': {'max_depth': [3, 5, 10], 'min_samples_split': [2, 10]},
    'XGBoost': {'n_estimators': [100], 'learning_rate': [0.1], 'max_depth': [3, 5]}
}
```

```python
cv_outer = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scenarios = {
    'Without SMOTE': (X_scaled, y),
    'With SMOTE': SMOTE(random_state=42).fit_resample(X_scaled, y),
    'With SMOTE and Sampling Strategy': SMOTE(sampling_strategy={1: int(850 * 0.5)}, random_state=42).fit_resample(X_scaled, y)
}

scenario_results = {}
for scenario_name, (X_scenario, y_scenario) in scenarios.items():
    model_performance = {}
    for name, model in models.items():
        outer_results = []
        for train_ix, test_ix in cv_outer.split(X_scenario, y_scenario):
            X_train, X_test = X_scenario[train_ix], X_scenario[test_ix]
            y_train, y_test = y_scenario[train_ix], y_scenario[test_ix]

            cv_inner = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
            grid = GridSearchCV(model, params[name], cv=cv_inner, scoring='accuracy')
            grid.fit(X_train, y_train)

            best_model = grid.best_estimator_
            y_pred = best_model.predict(X_test)
            acc = accuracy_score(y_test, y_pred)
            outer_results.append(acc)

        mean_accuracy = np.mean(outer_results)
        model_performance[name] = mean_accuracy
        print(f"Mean Accuracy for {name} in {scenario_name}: {mean_accuracy}")
        print("Confusion Matrix:")
        print(confusion_matrix(y_test, y_pred))

    scenario_results[scenario_name] = model_performance
    print(f"Results for {scenario_name}: {model_performance}")
for scenario_name, performance in scenario_results.items():
    plt.figure(figsize=(10, 6))
    model_names = list(performance.keys())
    accuracies = list(performance.values())
    sns.barplot(x=model_names, y=accuracies, palette='viridis')
    plt.title(f'Comparison of Model Performance - {scenario_name}')
    plt.xlabel('Models')
    plt.ylabel('Mean Accuracy')
    plt.ylim(0.7, 1)
    plt.show()
```
